```python
from matplotlib import pyplot as plt
import seaborn as sns
def _plot_series(series, series_name, series_index=0):
  palette = list(sns.palettes.mpl_palette('Dark2'))
  xs = series['Avg_Sr_clsr_time']
  ys = series['Avg_SR_open_days']

  plt.plot(xs, ys, label=series_name, color=palette[series_index % len(palette)])

fig, ax = plt.subplots(figsize=(10, 5.2), layout='constrained')
df_sorted = df.sort_values('Avg_Sr_clsr_time', ascending=True)
for i, (series_name, series) in enumerate(df_sorted.groupby('Mode_brand')):
  _plot_series(series, series_name, i)
  fig.legend(title='Mode_brand', bbox_to_anchor=(1, 1), loc='upper left')
sns.despine(fig=fig, ax=ax)
plt.xlabel('Avg_Sr_clsr_time')
_ = plt.ylabel('Avg_SR_open_days')
```
